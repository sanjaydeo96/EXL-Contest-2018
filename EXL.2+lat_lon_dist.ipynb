```python
import math
Nearest_shop=[]
Nearest_Distance=[]
for i in range(len(Ping_info)):
    shopper_lat=Ping_info.lat[i]
    shopper_lon=Ping_info.lng[i]
    disss=lambda lat1,lon1:acos(cos(math.radians(90-lat1)) *cos(math.radians(90-shopper_lat)) +sin(math.radians(90-lat1)) *sin(math.radians(90-shopper_lat)) *cos(math.radians(lon1-shopper_lon)))*6371/1.609

    Store_map['total_dist'] = list(map(disss,Store_map.latitude,Store_map.longitude))
    arg=np.argmin(Store_map[Store_map["Floor_Index"]==Ping_info.store_floor[i]]["total_dist"])
    Nearest_Distance.append(Store_map.total_dist[arg])
    Nearest_shop.append(Store_map.Store_Name[arg])
```
