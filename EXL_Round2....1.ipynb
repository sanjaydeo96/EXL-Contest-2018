```python
df_temp=df2[['Shopper_Id','labels']]
df_temp.columns=['Shopper_ID', 'labels']
```

```python
Demo_Ping_dup1_labeled=pd.merge(df1,df_temp,on='Shopper_ID',how='left')
Demo_Ping_dup1_labeled=Demo_Ping_dup1_labeled.drop('Unnamed: 0' ,axis=1)
df3=Demo_Ping_dup1_labeled
df3.head()
```

```python
# #Floor Wise Analysis
f=pd.crosstab(index=df3['floor'],columns=df3['labels'])
```

```python
f.plot(kind='bar',stacked=True,figsize=(8,8))
```
