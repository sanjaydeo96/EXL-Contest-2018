```python
Nearest_shop=[]
Nearest_Distance=[]
for i in range(len(Ping_info)):
    latt=np.square((Ping_info.lat[i])-(Store_map[Store_map['Floor_Index'] == Ping_info.store_floor[i]].latitude))
    lngg=np.square((Ping_info.lng[i])-(Store_map[Store_map['Floor_Index'] == Ping_info.store_floor[i]].longitude))
    min_dist=np.sqrt(latt+lngg)
    Nearest_Distance.append(min(min_dist))
    Nearest_shop.append(Store_map.Store_Name[np.argmin(min_dist)])    
```

```python
Ping_info['Nearest_shop']=Nearest_shop
Ping_info['Nearest_Distance']=Nearest_Distance
Ping_info.head(200)
```

```python
df_category=Category_map[['Store_Name','Fine_Category']]
di=dict((i,j) for i,j in zip(df_category.Store_Name,df_category.Fine_Category))
Ping_info['Fine_Category']=Ping_info.Nearest_shop.map(di)
```
